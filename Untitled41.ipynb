```python
from sklearn.metrics import adjusted_rand_score
from sklearn.utils import resample
np.random.seed(1234)

k_range = range(2, 9)  # Clusters from 2 to 8
nboot = 100  # Number of bootstrap samples
nrep = 10  # Number of repetitions per k

ari_scores = []

for k in k_range:
    cluster_labels = []
    
    for _ in range(nrep):
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=np.random.randint(0, 10000))
        labels = kmeans.fit_predict(MD_x)
        cluster_labels.append(labels)
    
    ari_values = []
    for _ in range(nboot):
        sample_indices = resample(np.arange(len(MD_x)))  # Bootstrap sample indices
        sample_labels_1 = np.array(cluster_labels[np.random.randint(0, nrep)])[sample_indices]
        sample_labels_2 = np.array(cluster_labels[np.random.randint(0, nrep)])[sample_indices]
        
        ari = adjusted_rand_score(sample_labels_1, sample_labels_2)
        ari_values.append(ari)
    
    ari_scores.append(np.mean(ari_values))

# Step 4: Plot ARI vs. number of clusters
plt.plot(k_range, ari_scores, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Adjusted Rand Index (ARI)")
plt.title("Bootstrap Stability Analysis of Clustering")
plt.show()
```

```python
kmeans_4 = KMeans(n_clusters=4, n_init=10, random_state=1234)
k4_labels = kmeans_4.fit_predict(MD_x)
sns.histplot(k4_labels, bins=np.arange(-0.5, 4.5, 1), discrete=True)
plt.xlim(0, 3)
plt.xlabel("Cluster Number")
plt.ylabel("Frequency")
plt.title("Cluster 4 Distribution")
plt.show()
```

```python
segment_stability = []
for _ in range(nboot):
    sample_indices = resample(np.arange(len(MD_x)))
    sample_labels_1 = np.array(k4_labels)[sample_indices]
    sample_labels_2 = np.array(k4_labels)[sample_indices]
    segment_stability.append(adjusted_rand_score(sample_labels_1, sample_labels_2))

plt.figure(figsize=(8, 5))
plt.plot(range(1, 5), segment_stability[:4], marker='o', linestyle='--')
plt.ylim(0, 1)
plt.xlabel("Segment Number")
plt.ylabel("Segment Stability (ARI)")
plt.title("Segment Stability Analysis")
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import adjusted_rand_score

np.random.seed(1234)

k_range = range(2, 9)
bic_scores = []
models = {}

for k in k_range:
    gmm = GaussianMixture(n_components=k, covariance_type="full", random_state=1234, n_init=10)
    gmm.fit(MD_x)
    bic_scores.append(gmm.bic(MD_x))
    models[k] = gmm  # Store models for later use

plt.plot(k_range, bic_scores, marker="o", linestyle="--")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("BIC Score")
plt.title("Model Selection using BIC")
plt.show()

best_k = 4
MD_m4 = models[best_k]
mixture_labels = MD_m4.predict(MD_x)

kmeans = KMeans(n_clusters=best_k, n_init=10, random_state=1234)
kmeans_labels = kmeans.fit_predict(MD_x)

contingency_table = pd.crosstab(kmeans_labels, mixture_labels, rownames=['KMeans'], colnames=['Mixture Model'])
print(contingency_table)
```

```python
import statsmodels.api as sm
import re

def extract_numeric(value):
    match = re.search(r'[-+]?\d+', str(value)) 
    return int(match.group()) if match else None 

data['Like'] = data['Like'].apply(extract_numeric)
data['Like.n'] = 6 - data['Like'].astype(int)

X = segment_variable.iloc[:, :10] 
y = data['Like.n'] 

X_with_const = sm.add_constant(X) 
reg_model = sm.OLS(y, X_with_const).fit()
print(reg_model.summary())  

np.random.seed(1234)
gmm = GaussianMixture(n_components=2, n_init=10, random_state=1234)
clusters = gmm.fit_predict(X)

data['Cluster'] = clusters

print(data['Cluster'].value_counts())

plt.hist(data['Cluster'], bins=2, edgecolor='black')
plt.xlabel("Cluster")
plt.ylabel("Count")
plt.title("Cluster Distribution (Gaussian Mixture Model)")
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.preprocessing import StandardScaler

MD_vclust = sch.linkage(MD_x.T, method='ward')

dendro = sch.dendrogram(MD_vclust, no_plot=True)
ordered_indices = dendro['leaves'][::-1]

sns.barplot(x=np.arange(len(ordered_indices)), y=ordered_indices, palette='coolwarm')
plt.xlabel("Clusters")
plt.ylabel("Ordered Indices")
plt.title("Cluster Bar Chart")
plt.show()

scaler = StandardScaler()
MD_x_scaled = scaler.fit_transform(MD_x)
MD_pca = PCA(n_components=2)
MD_proj = MD_pca.fit_transform(MD_x_scaled)

plt.scatter(MD_proj[:, 0], MD_proj[:, 1], c='b', label='Data Points')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Projection")
plt.legend()
plt.show()

plt.quiver([0, 0], [0, 0], MD_pca.components_[0, :], MD_pca.components_[1, :], angles='xy', scale_units='xy', scale=1, color='r')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Axes Projection")
plt.show()
```
